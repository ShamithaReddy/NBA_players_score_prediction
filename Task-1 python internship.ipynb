```python
y=list(data['pts'])
X=data.drop(columns=["player",'pts'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33)
```

```python
model=KNeighborsClassifier(n_neighbors=7)
model.fit(X_train,y_train)
```

```python
predictknn=model.predict(X_test)
```

```python
linear=LinearRegression()
linear.fit(X_train,y_train)
predictionlinear=linear.predict(X_test)
predictionlinear=list(predictionlinear)
```

```python
for i in range(len(predictionlinear)):
    predictionlinear[i]=int(predictionlinear[i])
```

```python
predictionlinear=np.array(predictionlinear)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,predictionlinear,normalize=False)
```
